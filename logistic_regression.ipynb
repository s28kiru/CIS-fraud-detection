```python

dummy_columns = ['ProductCD']

for i in range(0, len(dummy_columns)):
    colname = "df_" + str(i)
    train[dummy_columns[i]] = pd.Categorical(train[dummy_columns[i]])
    train[dummy_columns[i]] = train[dummy_columns[i]].cat.codes
    colname = pd.get_dummies(train[dummy_columns[i]])
    train = pd.concat([train, colname], axis = 1)
```
